```python
def plot_avg_hours_worked(df):
    """
    Plots the average hours worked per week, grouped by Gender and Work Location.

    Parameters:
    df (DataFrame): The input DataFrame with columns 'Gender', 'Work_Location', 
	and 'Hours_Worked_Per_Week'.
    """

    # Group data by 'Gender' and 'Work_Location', calculating the mean of 
	# 'Hours_Worked_Per_Week' for each group
    data = df.groupby(["Gender", "Work_Location"])["Hours_Worked_Per_Week"].mean()

    # Reshape the grouped data so that 'Work_Location' categories become columns,
    # making it easier to plot as a bar chart
    data_df = data.unstack()

    # Plot the data as a bar chart with specified figure size
    data_df.plot(kind='bar', figsize=(7, 5))

    # Adjust the legend to appear outside the main plot area for clarity
    plt.legend(bbox_to_anchor=(1.21, 1), loc='upper right')

    # Set the x-axis labels to be horizontal (rotation=0) for better readability
    plt.xticks(rotation=0)

    # Label the y-axis to indicate the data represents "Hours Worked Per Week"
    plt.ylabel("Hours Worked Per Week")

    # Set the title of the plot to describe the chart's information
    plt.title("Average Hours Worked Per Week")

    # Display the plot
    plt.show()

# Call the plot function
plot_avg_hours_worked(df)
```

```python
def plot_avg_hours_worked_line(df):
    """
    Plots the average hours worked per week as a line chart, grouped by Gender 
	and Region.

    Parameters:
    df (DataFrame): The input DataFrame with columns 'Gender', 'Region', and 
	'Hours_Worked_Per_Week'.
    """

    # Group the data by 'Gender' and 'Region', calculating the mean of 
	# 'Hours_Worked_Per_Week' for each group
    df_data = df.groupby(["Gender", "Region"])["Hours_Worked_Per_Week"].mean()

    # Reshape the grouped data so that 'Region' categories become columns,
    # making it easier to plot as a line chart
    df_data = df_data.unstack()

    # Plot the data as a line chart
    df_data.plot(kind='line')

    # Label the y-axis to indicate the data represents "Hours Worked"
    plt.ylabel("Hours Worked")

    # Set the title of the plot to describe the chart's information
    plt.title("Average Hours Worked per Week")

    # Display the plot
    plt.show()

# Call the plot function
plot_avg_hours_worked_line(df)
```

```python
def plot_boxplot_experience_meetings_balance(df):
    """
    Plots a boxplot for 'Years_of_Experience', 'Number_of_Virtual_Meetings', and
    'Work_Life_Balance_Rating' columns in the DataFrame.

    Parameters:
    df (DataFrame): The input DataFrame with columns 'Years_of_Experience', 
                    'Number_of_Virtual_Meetings', and 'Work_Life_Balance_Rating'.
    """

    # Set the figure size for the plot to 12 inches wide by 6 inches tall
    plt.figure(figsize=(12, 6))

    # Create a boxplot for the specified columns to show the distribution of values
    df.boxplot(column=["Years_of_Experience", "Number_of_Virtual_Meetings", 
                       "Work_Life_Balance_Rating"])

    # Set the title of the plot to 'Stress Level by Job Role'
    plt.title('Stress Level by Job Role')

    # Label the y-axis as 'Value' to represent the metric being measured
    plt.ylabel('Value')

    # Label the x-axis as 'Group' to indicate each group in the boxplot
    plt.xlabel('Group')

    # Display the plot
    plt.show()

# Call the plot function
plot_boxplot_experience_meetings_balance(df)
```

```python
def plot_correlation_heatmap(df):
    """
    Plots a correlation heatmap for selected columns related to work-life balance, 
    social isolation, experience, and hours worked.

    Parameters:
    df (DataFrame): The input DataFrame with columns 'Work_Life_Balance_Rating', 
                    'Social_Isolation_Rating', 'Years_of_Experience', 
                    and 'Hours_Worked_Per_Week'.
    """

    # Select the relevant columns for the correlation matrix
    df_corr = df[["Work_Life_Balance_Rating", "Social_Isolation_Rating", 
                  "Years_of_Experience", "Hours_Worked_Per_Week"]]

    # Compute the correlation matrix for the selected columns
    correlation_matrix = df_corr.corr()

    # Set the figure size for the heatmap to 6 inches by 4 inches
    plt.figure(figsize=(6, 4))

    # Plot the correlation matrix as a heatmap with annotations
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', 
                square=True, cbar=True)

    # Set the title of the plot to 'Correlation Matrix'
    plt.title('Correlation Matrix')

    # Display the plot
    plt.show()

# Call the plot function
plot_correlation_heatmap(df)
```
